## Feature Engineering

I would like to know if the numbers of nights that the user is looking to stay at the hotel, has any relationship with the hotel cluster
For this, I need to convert columns check-in and check-out datetime format in integer. As weell as date_time column needs to be converted to an integer

```python
from datetime import datetime
```

```python
df_train['user_date'] = [str(d).split(' ')[0] for d in df_train['date_time']]
df_train['user_time'] = [str(t).split(' ')[1] for t in df_train['date_time']]
df_train['user_date'] = [datetime.strptime(d,'%Y-%m-%d') for d in df_train['user_date']]
df_train['user_time'] = [datetime.strptime(t,'%H:%M:%S') for t in df_train['user_time']]
```

```python
df_train['srch_co'] =  pd.to_datetime(df_train['srch_co'], format = '%Y-%m-%d')
df_train['srch_ci'] =  pd.to_datetime(df_train['srch_ci'], format = '%Y-%m-%d')
```

```python
df_train['srch_year_co'] = [(x.year) for x in df_train['srch_co']]
df_train['srch_month_co'] = [(x.month) for x in df_train['srch_co']]
df_train['srch_day_co'] = [(x.day) for x in df_train['srch_co']]
df_train['srch_year_ci'] = [(x.year) for x in df_train['srch_ci']]
df_train['srch_month_ci'] = [(x.month) for x in df_train['srch_ci']]
df_train['srch_day_ci'] = [(x.day) for x in df_train['srch_ci']]
```

```python
df_train['user_year'] = [(x.year) for x in df_train['user_date']]
df_train['user_month'] = [(x.month) for x in df_train['user_date']]
df_train['user_day'] = [(x.day) for x in df_train['user_date']]
```

I am not taking in consideration seconds in the user_time column for this study

### Missing values

```python
#Filling up missing values: destination_distance, using the mean
df_train['orig_destination_distance'].fillna(df_train['orig_destination_distance'].mean(), inplace=True)
```

```python
#Filling up missing values: srch_co and srch_ci, using the most repetitive value of each
df_train['srch_year_co'].fillna(df_train.srch_year_co.mode(), inplace=True)
df_train['srch_month_co'].fillna(df_train.srch_month_co.mode(), inplace=True)
df_train['srch_day_co'].fillna(df_train.srch_day_co.mode(), inplace=True)
df_train['srch_year_ci'].fillna(df_train.srch_year_ci.mode(), inplace=True)
df_train['srch_month_ci'].fillna(df_train.srch_month_ci.mode(), inplace=True)
df_train['srch_day_ci'].fillna(df_train.srch_day_ci.mode(), inplace=True)
```

```python
#Dropping timestamp coumns that were already processed
df_train =df_train.drop('srch_co', axis = 1)
df_train =df_train.drop('srch_ci', axis = 1)
df_train =df_train.drop('date_time', axis = 1)
```

```python
df_train =df_train.drop('user_time', axis=1)
```

```python
df_train =df_train.drop('user_date', axis = 1)
```
